```python
import re
import numpy as np
```

```python
# read the text file
with open('glossary.txt', 'r', encoding="ISO-8859-1") as f:
    lines = f.readlines()
    
tmp=[]
for line in lines:
    parts=re.split('\n',line)
    for part in parts:
        if part:
            tmp.append(part)
lines=tmp
```

```python
# remove the lines with 'Pages'
glossary=[re.sub(r'Page \d+ Â¥?','',line) for line in lines]

# extract phase before ':'
tmp=[]
pattern=re.compile(r'^.*(?=:)')
pattern_abbre=re.compile(
    r'(?<=abbreviated as\s).*?(?=[\)\.\,])') # extract the word after abbreviated as
pattern_abbre_sub=re.compile(
    r'abbreviated as\s.*?[\)\.\,]\s?')
for line in glossary:
    line=re.split(r'\.\s+',line)
    for sentence in line:
        if sentence:
            
            phase=pattern_abbre.findall(sentence)
            if phase: tmp.append(phase)
            sentence=re.sub(pattern_abbre_sub,'',sentence)
            
            phase=pattern.findall(sentence)
            if phase: tmp.append(phase)
            
glossary=tmp
# glossary=[word for word in glossary if word]
```

```python
# for phases having special symbol (e.g. - , \, /, \,) or ' or ' in them,
# need further breaking down of them

tmp=[]
split_pattern=re.compile(r'\s-\s|\\|\s\/\s|\sor\s')
for phases in glossary:
    words=split_pattern.split(phases[0])
    for word in words:
        if ':' in word:
            # in case there is missing filtering of ":"
            tmp.append(re.split(':',word)[0])
        elif word:
            tmp.append(word)

# further extract the word if there is brackets in them 
# (or there are words 'abbreviated as')
tmp2=[]
for word in tmp:
    
    if '(' in word: 
        pattern1=re.compile(r'.*(?=\s+\()')
        tmp2.append(pattern1.findall(word)[0])
        pattern2=re.compile(r'(?<=\().*(?=\))')
        if pattern2.findall(word):
            if ',' in pattern2.findall(word)[0]:
                tmp2.append(re.split(',',pattern2.findall(word)[0])[0])
            else:
                tmp2.append(pattern2.findall(word)[0])
    elif ',' in word:
        split_tmp=re.split(',',word)
        for split_word in split_tmp:
            tmp2.append(split_word)
    else:
        tmp2.append(word)

cleaned_glossary=[tmp_word.strip() for tmp_word in tmp2]
# avoid repetition of phases
cleaned_glossary=list(set(cleaned_glossary))
```
